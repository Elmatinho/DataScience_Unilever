## Ejercicio
Hacer un programa que pida numero de estudiantes a registrar, luego de esto pida la calificación para 4 asignaturas(matematicas, ciencias, historia, ingles), me lo van a transformar a un DF. Van a filtrar el df, con los estudiantes que hayan reprobado al menos una asignatura. La escala de 0-6 reprobado, de 7 a 10 aprobado.

```python
alumno = []
asignatura1=[]
asignatura2=[]
asignatura3=[]
asignatura4=[]
alumnos = {}
cant_alumnos=int(input("cantidad de estudiantes"))

for h in range(cant_alumnos):
  alumnoin = input("\n Ingrese el nombre del alumno: ")
  alumno.append(alumnoin)
  asignatura1in=float(input("\n Ingrese la nota de mates: "))
  asignatura1.append(asignatura1in)
  asignatura2in=float(input("\n Ingrese la nota de ciencia: "))
  asignatura2.append(asignatura2in)
  asignatura3in=float(input("\n Ingrese la nota de historia: "))
  asignatura3.append(asignatura3in)
  asignatura4in=float(input("\n Ingrese la nota de ingles: "))
  asignatura4.append(asignatura4in)
alumnos['nombre'] = alumno
alumnos['mates'] = asignatura1
alumnos['ciencias'] = asignatura2
alumnos['historia'] = asignatura3
alumnos['ingles'] = asignatura4


notillas=pd.DataFrame(alumnos)
notillas

#Falto los filtros


```

# Ejercicio
Crear un control de contabilidad para una tienda, en primera instancia se debe crear un constructor de inventario, donde se van a recibir productos por entrada hasta que se coloque la palabra salir. A cada uno de los productos se le va a solicitar la siguiente informacion (ID, CANTIDAD EN EXISTENCIA, CANTIDAD VENDIDA, ITEMS PERDIDOS, TOTAL GANANCIA). Una vez que finaliza el constructor de inventario, se debe mostrar el dataframe construido y un menu con 5 filtros diferentes a implementar que seran elegidos con los numeros del 1 al 5, cuando se digite 0, el programa termina.

```python
idlista=[]
cantidadexlista=[]
cantidadvendidalista=[]
items_pedidos_lista=[]
total_ganacia_lista=[]
constructor={}
ingresar=""
while ingresar != "salir":
    if ingresar!="salir":
        id=int(input("ingrese el id"))
        idlista.append(id)
        cantidad=int(input("ingrese la cantidad en existencia\n"))
        cantidadexlista.append(cantidad)
        vendido=int(input("ingrese la cantidad vendida\n"))
        cantidadvendidalista.append(vendido)
        items=int(input("ingrese los itemes perdidos\n"))
        items_pedidos_lista.append(items)
        ganacias=int(input("ingrese las ganacias\n"))
        total_ganacia_lista.append(ganacias)
        ingresar=input("Desea agregar un producto? si no lo desea escriba:\n Salir\n")
constructor['ID']=idlista
constructor['Cantidades en existencia']=cantidadexlista
constructor['Cantidades vendidas']=cantidadvendidalista
constructor['Itemes perdidos']=items_pedidos_lista
constructor['total ganacia']=total_ganacia_lista

inventario=pd.DataFrame(constructor)
inventario

```
