## Importación y cambio de nombre de las columnas

```python
import pandas as pd
filename='/Users/matiasinostrozacifuentes/Desktop/python_intro/DataScience_Unilever/Semana_2/Data/CalIt2.data.csv'
df=pd.read_csv(filename,sep=';',names=['ID_Flujo','Fecha','Hora','Conteo'])
df
```

## Seleccionando datos

```python
#1
ID_filtrado=df['ID_Flujo']==7
Fecha_filtrada=df['Fecha']=='07/24/05'
df2=df.loc[Fecha_filtrada & ID_filtrado, :]
df2
```

```python
#2
df3=df2['Conteo']
df3=df3.sort_values(ascending=False)
TOP10=df3.head(10)
```

```python
TOP10
```

```python
TOP10.apply(lambda num: num-TOP10.min()).mean()
```
