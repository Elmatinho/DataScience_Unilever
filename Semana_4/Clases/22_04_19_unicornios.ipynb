## 3. preprocesamiento

```python
df["Valuation ($B)"]=df["Valuation ($B)"].str.replace("$","")
df['Valuation ($B)']=pd.to_numeric(df['Valuation ($B)'],errors='coerce')
```

```python
df.info()
```

Cambiando de tipo de dato a date joined

```python
df['Date Joined'] = pd.to_datetime(df['Date Joined'])
```

```python
df.info()
```

## 4. Analisis


```python
#1 top 5

top5=df.groupby('Company')['Valuation ($B)'].sum().sort_values().tail(5)
```

```python
#figura 1
fig = plt.figure(dpi=200)
 #para ajustar los gráficos
plt.style.use('seaborn')
ax1 = fig.add_subplot(2,2,1)
ax1.barh(top5.index,top5.values)
ax1.set_xlabel('Valorización ($B)')
ax1.set_ylabel('Compañias')
ax1.set_title('Top 5 de compañias con mayor valorización',fontsize=10)
plt.xticks(rotation=90)


ax2=fig.add_subplot(2,2,2)
ax2.pie(top5.values,autopct="%.1f%%")
ax2.legend(top5.index,loc=[1.02,0.1],fontsize=7)
ax2.set_title('Top5 de compañias con mayor valorización',fontsize=10)
fig.tight_layout()
```

## 4.3 Ciudad 

```python
ciudad=df.groupby('City')['Company'].count().sort_values()
```

```python
unicornios=pd.DataFrame(ciudad)
```

```python
unicornios.index=range(256)
unicornios
```

```python
Categoria=[]
for i in unicornios['Company']:
    if i==0:
        Categoria.append('No hay Unicornios')
    elif i>=1 and i<=5:
        Categoria.append('Pocos unicornios')
    elif i>5 and i<=10:
        Categoria.append('Cantidad normal de Unicornios')
    elif i>10 and i<=60:
        Categoria.append('Hay muchos unicornios')
    else:
        Categoria.append('Extremadamente muchos unicornios')
unicornios['Categoria']=Categoria
```

```python
unicornios
```

```python
unicornios.groupby('Categoria')['Company'].mean()
```

```python
bar=sns.barplot(y=unicornios.groupby('Categoria')['Company'].mean().index,x=unicornios.groupby('Categoria')['Company'].mean().values)
plt.xticks(rotation=90,fontsize=15)
```
