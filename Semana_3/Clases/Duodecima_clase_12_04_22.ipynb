```python
from numpy import sin,arange,pi,cos
x=arange(0.0,2*pi,0.1)
z=cos(x)
y=sin(x)
plt.plot(x,y,'b-',x,z,'r-')
plt.grid(axis='y')
plt.axis([0,6,-1,1])
plt.show()
```

```python
bancos_filename='/Users/matiasinostrozacifuentes/Desktop/python_intro/DataScience_Unilever/Semana_3/Data/bancos.csv'
df_bancos=pd.read_csv(bancos_filename)
df_bancos
```

```python
df_bancos.info()
```

```python
df_bancos['Empresa'].unique()
```

```python
df_bancos=df_bancos.sort_values(by='Fecha')
```

```python
df_bancos['Promedio']=(df_bancos['Máximo']+df_bancos['Mínimo'])/2
df_bancos
```

```python
filtro_santander=df_bancos.loc[df_bancos['Empresa']=='Banco Santander', :]
filtro_sabadell=df_bancos.loc[df_bancos['Empresa']=='Banco Sabadell', :]
filtro_bankinter=df_bancos.loc[df_bancos['Empresa']=='Bankinter', :]
filtro_bbva=df_bancos.loc[df_bancos['Empresa']=='BBVA', :]
filtro_caixabank=df_bancos.loc[df_bancos['Empresa']=='Caixabank', :]
```

```python
filtro_bankinter
```

```python
plt.figure(dpi=100)
plt.plot(filtro_sabadell['Fecha'],filtro_sabadell['Cierre'],'r.-',filtro_santander['Fecha'],filtro_santander['Cierre'],'b--',filtro_bankinter['Fecha'],filtro_bankinter['Cierre'],'go-',filtro_bbva['Fecha'],filtro_bbva['Cierre'],'c',
filtro_caixabank['Fecha'],filtro_caixabank['Cierre'],'yo-')
plt.show()
```

```python
plt.figure(dpi=100)
plt.plot(filtro_sabadell['Fecha'].dt.day,filtro_sabadell['Promedio'],'r.-',filtro_santander['Fecha'].dt.day,filtro_santander['Promedio'],'b--',filtro_bankinter['Fecha'].dt.day,filtro_bankinter['Promedio'],'go-',filtro_bbva['Fecha'].dt.day,filtro_bbva['Promedio'],'c',
filtro_caixabank['Fecha'].dt.day,filtro_caixabank['Promedio'],'yo-')
plt.show()
```
