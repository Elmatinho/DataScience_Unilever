```python
data['date'].unique()
```

```python
new_date=[]
for i in data['date']:
    if "T000000" in i:
        i=i.replace("T000000","")
        i=pd.to_datetime(i)
        new_date.append(i)
data['Fecha']=new_date
```

```python
data.drop(columns=['date'],inplace=True)
```

separando la data

```python
X = data.drop(columns=['Fecha','price'])
y = data.price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

Bagging tree

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
bag_rgrs = BaggingRegressor(
    DecisionTreeRegressor(), n_estimators=300,
    max_samples=300, bootstrap=True
)

bag_rgrs.fit(X_train, y_train)
y_pred = bag_rgrs.predict(X_test)
print(y_pred)
print(y_test)
bag_rgrs.score(X_test, y_test)
```

Con random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rndf_rgrs=RandomForestRegressor(n_estimators=300)
rndf_rgrs.fit(X_train, y_train)
y_pred = rndf_rgrs.predict(X_test)

rndf_rgrs.score(X_test, y_test)
```

Claramente el numero de estimadores es unfactor clave en ambos algortimos, pero en los bagging regressor el numero de muestras tambien esextremadamente representativo en la precision del modelo
