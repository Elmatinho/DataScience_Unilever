Histograma de los posedores de perros

```python
hist1=data_GP['Percentage of Dog Owners'].hist()
hist2=data_GP['Percentage of Cat Owners'].hist()
```

```python
plt.grid(True)
hist1=sns.histplot(data=data_GP,x='Percentage of Dog Owners',kde=True)
hist2=sns.histplot(data=data_GP,x='Percentage of Cat Owners',color='Orange',kde=True)
plt.legend(['Dogs','Cats'])
hist1.set_title('Distribucin de porcentaje de dueos de mascotas')
hist1.set_xlabel('Porcentaje de dueos')
```

Como se puede apreciar, el porcentaje de dueos de gatos su distribucin se concentra ms hacia la izquierda, donde su media claramente recae en el 30%, mientras que los dueos de perros son en mayor cantidad que el de los gatos, donde su media se situa en 35%

Primero se debe hacer un melt para tener todos los valores de perros y los de gatos

```python
data_GP_melted=pd.melt(data_GP)
data_GP_melted
```

Haciendo filtros

```python
Filtrado=data_GP_melted.loc[(data_GP_melted['variable']=='Mean Number of Dogs per household') | (data_GP_melted['variable']=='Mean Number of Cats'),:]
```

```python
f, ax = plt.subplots(figsize=(7, 6))

ax=sns.boxplot(x='variable',y='value',data=Filtrado)
ax.set_xticklabels(['Perros','Gatos'])
ax.set_xlabel('Animal')
ax.set_ylabel('Valores')

```
