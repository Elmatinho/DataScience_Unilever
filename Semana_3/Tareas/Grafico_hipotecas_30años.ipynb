Importando librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargando datos

```python
filename1="/Users/matiasinostrozacifuentes/Desktop/python_intro/DataScienceUnilever2022/Track 1/Clases/2-Pandas/data/mortgages.csv"
df1=pd.read_csv(filename1)
df1
```

Filtrando solo por 30 años

```python
df1=df1.loc[df1['Mortgage Name']=='30 Year', :]
df1
```

```python
df1.info()
```

```python
df_003=df1.loc[df1['Interest Rate']==0.03]
df_005=df1.loc[df1['Interest Rate']==0.05]
df_005
```

```python
plt.figure(dpi=100)
plt.plot(df_003['Month'],df_003['Interest Paid'].cumsum(),'c',df_005['Month'],df_005['Interest Paid'].cumsum())
plt.ylabel('Dinero')
plt.xlabel('Meses')
plt.title('Comparación entre pago de intereses de hipotecas')
plt.legend(['Interes al 3%','Interes al 5%'])
plt.grid(axis='both')
plt.show()
```
